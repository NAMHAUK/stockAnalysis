```python
import csv

# Load Data
file_path = './drive/Shareddrives/CS360/madcamp-3-week/data/'
file_path = file_path + '삼성전자_일별데이터.csv'

# Save File
_file = open(file_path, 'r', encoding = 'cp949')
file_rdr = csv.reader(_file)

date_lst = []



file_dict = dict()
for line in file_rdr:
  if line[0] == '일자':
    continue


  if line[0][2:-3] in ['07-11', '07-12', '08-01', '08-02', '08-03', '08-04', '08-05', '08-06', '08-07', 
                       '08-08', '08-09', '08-10', '08-11', '11-05', '11-06', '11-07', '11-08', '11-09']:
    file_dict[line[0]] = {
        '시가' : int(line[1]),
        '종가' : int(line[2]),
        '기관계' : int(line[3]),
        '금융투자' : int(line[4])
    }
    date_lst.append(line[0])
  
  elif line[0][2:4] == '18':
    if line[0][5:7] != '01':
      file_dict[line[0]] = {
          '시가' : int(line[1]),
          '종가' : int(line[2]),
          '기관계' : int(line[3]),
          '금융투자' : int(line[4])
      }
      date_lst.append(line[0])
  
  elif line[0][2:4] == '19':
    file_dict[line[0]] = {
        '시가' : int(line[1]),
        '종가' : int(line[2]),
        '기관계' : int(line[3]),
        '금융투자' : int(line[4])
    }
    date_lst.append(line[0])
  
  elif line[0][2:-3] in ['20-01', '20-02', '20-03']:
    if line[0][-2] == '2' or '3':
      file_dict[line[0]] = {
        '시가' : int(line[1]),
        '종가' : int(line[2]),
        '기관계' : int(line[3]),
        '금융투자' : int(line[4])
      }
      date_lst.append(line[0])
  
  elif line[0][2:-3] in ['21-07', '21-08', '21-09', '21-10', '21-11', '21-12']:
    file_dict[line[0]] = {
      '시가' : int(line[1]),
      '종가' : int(line[2]),
      '기관계' : int(line[3]),
      '금융투자' : int(line[4])
    }
    date_lst.append(line[0])
  
  elif line[0][2:4] == '22':
    file_dict[line[0]] = {
      '시가' : int(line[1]),
      '종가' : int(line[2]),
      '기관계' : int(line[3]),
      '금융투자' : int(line[4])
    }
    date_lst.append(line[0])

date_lst.reverse()
```

```python
tgt_cnt = 0
tgt_date_lst = []
open_open_win, open_open_lose = [], []
open_close_win, open_close_lose = [], []
close_open_win, close_open_lose = [], []
close_close_win, close_close_lose = [], []
for i in range(len(date_lst)):
  if i == len(date_lst) - 1:
    break
  
  curr = date_lst[i]
  next = date_lst[i + 1]

  curr_open = file_dict[curr]['시가']
  curr_close = file_dict[curr]['종가']
  curr_vol = file_dict[curr]['기관계']
  curr_tgt = file_dict[curr]['금융투자']

  next_open = file_dict[next]['시가']
  next_close = file_dict[next]['종가']

  if curr_tgt < 0 or curr_vol < 0:
    continue
  
  if curr_tgt / curr_vol >= 0.3:
    tgt_cnt += 1
    tgt_date_lst.append(curr)

    open_open_gap = next_open - curr_open
    open_open_per = (open_open_gap / curr_open) * 100
    if open_open_per > 0 :
      open_open_win.append(open_open_per)
    else:
      open_open_lose.append(open_open_per)

    open_close_gap = next_open - curr_close
    open_close_per = (open_close_gap / curr_close) * 100
    if open_close_per > 0:
      open_close_win.append(open_close_per)
    else:
      open_close_lose.append(open_close_per)

    close_open_gap = next_close - curr_open
    close_open_per = (close_open_gap / curr_open) * 100
    if close_open_per > 0:
      close_open_win.append(close_open_per)
    else:
      close_open_lose.append(close_open_per)

    close_close_gap = next_close - curr_close
    close_close_per = (close_close_gap / curr_close) * 100
    if close_close_per > 0:
      close_close_win.append(close_close_per)
    else:
      close_close_lose.append(close_close_per)

print(tgt_date_lst)
```

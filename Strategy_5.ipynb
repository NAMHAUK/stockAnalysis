```python
import pandas as pd
import numpy as np

results1 = dict() # 배당락일 공매도
results2 = dict() # 배당락일 현물매수

def make_date_in_calendar(date, calendar):
  if date not in calendar:
    day = int(date[-2:])
    day -= 1
    date = date[:-2] + str(day)
    if date in calendar:
      return date
    else:
      return make_date_in_calendar(date, calendar)
  else:
    return date

def get_results(stock_name):

  print(f'[{stock_name}]')

  alloc_file_path = f'./drive/Shareddrives/CS360/madcamp-3-week/data/strategy-6/{stock_name}_배당일정.csv'
  price_file_path = f'./drive/Shareddrives/CS360/madcamp-3-week/data/strategy-6/{stock_name}_일별주가.csv'
  alloc_file = pd.read_csv(alloc_file_path, encoding = 'CP949', thousands = ',')
  price_file = pd.read_csv(price_file_path, encoding = 'CP949', thousands = ',')

  alloc_date = list(alloc_file[(alloc_file['현금배당금'] > 0) & (alloc_file['주식종류'] == '보통')]['기준일'])

  # 공매도 금지 기간 필터링
  for date in alloc_date.copy(): 
    if date.startswith('2020') or date.startswith('2021/03'):
      alloc_date.remove(date)

  # 거래일 아닌 일자 거래일로 만들기
  for i in range(len(alloc_date)):
    calendar = list(price_file['일자'])
    alloc_date[i] = make_date_in_calendar(alloc_date[i], calendar)

  data1 = price_file.iloc[np.array(price_file.index[price_file['일자'].isin(alloc_date)]) + 1] 
  data2 = price_file.iloc[np.array(price_file.index[price_file['일자'].isin(alloc_date)]) + 2] 

  # 1.배당락일 공매도
  data1_open = np.array(data1['시가'])
  data1_close = np.array(data1['종가'])
  data1_profit = 100 * (data1_open - data1_close) / data1_open

  # 2.배당락전일 현물매수
  data2_open = np.array(data2['시가'])
  data2_close = np.array(data2['종가'])
  data2_profit = 100 * (data2_close - data2_open) / data2_open

  results1[stock_name] = {
      'total' : len(alloc_date),
      '승리' : np.sum(np.where(data1_profit > 0, 1, 0)),
      '패배' : np.sum(np.where(data1_profit <= 0, 1, 0)),
      '평균수익률' : np.average(np.where(data1_profit > 0, data1_profit, 0)),
      '평균손실률' : np.average(np.where(data1_profit <= 0, data1_profit, 0)),
  }

  results2[stock_name] = {
      'total' : len(alloc_date),
      '승리' : np.sum(np.where(data2_profit > 0, 1, 0)),
      '패배' : np.sum(np.where(data2_profit <= 0, 1, 0)),
      '평균수익률' : np.average(np.where(data2_profit > 0, data2_profit, 0)),
      '평균손실률' : np.average(np.where(data2_profit <= 0, data2_profit, 0)),
  }
  
  print('-배당락일 공매도')
  print(results1[stock_name])
  print('-배당락전일 현물매수')
  print(results2[stock_name])
  print()


  


stock_list = ['삼성증권', 'NH투자증권', '한국금융지주', '이베스트투자증권', '효성', '하나금융지주', '기업은행', '금호석유', '한국가스공사', '우리금융지주', 'LX인터내셔널', '삼성카드', 'POSCO홀딩스', '쌍용C&E']
for stock_name in stock_list:
  get_results(stock_name)

```

```python
# 배당락일 공매도
data1 = []
cols1 = ['종목명', '승리', '패배', '승률(%)', '패율(%)', '평균수익률(%)', '평균손실률(%)', '기대수익률(%)']

for stock_name in stock_list:
  nw = results1[stock_name]['승리']
  nl = results1[stock_name]['패배']
  w = nw / results1[stock_name]['total']
  l = nl / results1[stock_name]['total']
  wp = results1[stock_name]['평균수익률']
  lp = results1[stock_name]['평균손실률']
  res = w * wp + l * lp

  d = [stock_name, nw, nl, f"{w*100:.2f}", f"{l*100:.2f}", f"{wp:.2f}", f"{lp:.2f}", f"{res:.2f}"]
  data1.append(d)

DF1 = pd.DataFrame(data = data1, columns=cols1)
DF1
```

```python
# 배당락전일 현물매수
data2 = []
cols2 = ['종목명', '승리', '패배', '승률(%)', '패율(%)', '평균수익률(%)', '평균손실률(%)', '기대수익률(%)']

for stock_name in stock_list:
  nw = results2[stock_name]['승리']
  nl = results2[stock_name]['패배']
  w = nw / results2[stock_name]['total']
  l = nl / results2[stock_name]['total']
  wp = results2[stock_name]['평균수익률']
  lp = results2[stock_name]['평균손실률']
  res = w * wp + l * lp

  d = [stock_name, nw, nl, f"{w*100:.2f}", f"{l*100:.2f}", f"{wp:.2f}", f"{lp:.2f}", f"{res:.2f}"]
  data2.append(d)

DF2 = pd.DataFrame(data = data2, columns=cols2)
DF2
```
